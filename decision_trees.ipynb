```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.tree import DecisionTreeClassifier
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_wine
from sklearn.metrics import accuracy_score
```

```python
iris = load_wine()
X = pd.DataFrame(iris.data, columns=iris.feature_names)
y = iris.target
```

```python
# y = y.tolist()
# for i in range(len(y)):
#     class_arr = [0] * 3
#     class_arr[y[i]] = 1
#     y[i] = class_arr
# y = np.array(y)
# y.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=10)
```

```python
clf = DecisionTreeClassifier()
clf.fit(X_train, y_train)
```

```python
predictions = clf.predict(X_test)
predictions.shape, y_test.shape
```

```python
accuracy_score(predictions, y_test)
```

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(25,10))
a = plot_tree(clf, 
              feature_names=iris.feature_names, 
              class_names=iris.target_names, 
              filled=True, 
              rounded=True, 
              fontsize=14)
```

# CovType Processing

```python
from sklearn.datasets import fetch_covtype
```

```python
covtype = fetch_covtype()
```

```python
X_c = pd.DataFrame(covtype['data'], columns=covtype.feature_names)
y_c = covtype.target
```

```python
set(y_c)
```

```python
dec = [0,1,2,3,4,5,6]
for dec_num in dec:
    for i in range(len(y_c)):
        if y_c[i] == dec_num + 1:
            y_c[i] = dec_num
set(y_c)
```

```python
X_train_c, X_test_c, y_train_c, y_test_c = train_test_split(X_c, y_c, random_state=10)
```

```python
X_train_c
```

```python
train_set_c = pd.DataFrame(X_train_c)
train_set_c['Cover_Type'] = y_train_c
```

```python
test_set_c= pd.DataFrame(X_test_c)
test_set_c['Cover_Type'] = y_test_c
```

```python
train_set_c
```

```python
test_set_c.to_csv('covtype_test.csv', sep = ',', index = False)
train_set_c.to_csv('covtype_train.csv', sep = ',', index = False)
```
