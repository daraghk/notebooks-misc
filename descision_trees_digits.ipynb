## Digits Dataset

```python
from sklearn import datasets
import pandas as pd 

digits_train = pd.read_csv('/Users/daraghking/Documents/Thesis/Code/Rust/grad_boost_mcc/tree_learner/common/data-files/digits_train.csv')
digits_test = pd.read_csv('/Users/daraghking/Documents/Thesis/Code/Rust/grad_boost_mcc/tree_learner/common/data-files/digits_test.csv')

digits_train
```

```python
X_train = digits_train.drop('digit', 1)
y_train = digits_train['digit']
y_train = pd.get_dummies(y_train)

X_test = digits_test.drop('digit', 1)
y_test = digits_test['digit']
y_test = pd.get_dummies(y_test)
```

```python
from sklearn.multioutput import MultiOutputClassifier
from sklearn.ensemble import GradientBoostingClassifier

import time

start = time.time()
params = {'n_estimators': 100, 'max_depth': 12, 'criterion' :'mse', 'learning_rate' : 0.1}
estimator = MultiOutputClassifier(GradientBoostingClassifier(**params))
estimator.fit(X_train, y_train)
end = time.time()
diff = end - start
print('Time elapsed', diff)
```

```python
from sklearn.metrics import accuracy_score
y_pred = estimator.predict(X_test)
accuracy_score(y_test, y_pred)
```

## MNIST using sk-learn

```python
from sklearn.datasets import fetch_openml
mnist = fetch_openml('mnist_784', version=1, cache=True)
```

```python
X_mnist = mnist.data.astype('float64')
y_mnist = mnist.target
```

```python
X_mnist.shape, y_mnist.shape
```

```python
X_train_mnist, X_test_mnist, y_train_mnist, y_test_mnist = train_test_split(X_mnist, y_mnist, random_state=10, test_size=0.1)
y_train_mnist_one_hot = pd.get_dummies(y_train_mnist)
y_test_mnist_one_hot = pd.get_dummies(y_test_mnist)
```

## LGBM MNIST

```python
import lightgbm as lgb
from sklearn.multioutput import MultiOutputClassifier
import time
```

```python
from lightgbm import LGBMClassifier
```

```python
lgbm_classifier = LGBMClassifier()
```

```python
lgb_start = time.time()
params = {'n_estimators': 100, 'max_depth': 3, 'learning_rate' : 0.5, 'num_leaves': 1000, 'min_child_samples': 1}
lgb_estimator = MultiOutputClassifier(LGBMClassifier(**params))
lgb_estimator.fit(X_train_mnist, y_train_mnist_one_hot)
lgb_end = time.time()
lgb_time_diff = lgb_end - lgb_start
print('Time elapsed', lgb_time_diff)
```

```python
y_predm = lgb_estimator.predict(X_test_mnist)
accuracy_score(y_test_mnist_one_hot, y_predm)
```
